# Membuat plot per hari

```python
lantai = 6
dfOpen = data2018[lantai-1].copy()
```

```python
df = dfOpen
# print(df.columns)
kolom = df.columns
kolomBaru = []
for i in kolom:
    if i[-2] == "W" or i == "Date" or i == "Hari" or i == "Tanggal":
        kolomBaru.append(i)
```

```python
kolomBaru
```

```python
kolomBaruSum = kolomBaru[1:]
```

```python
df = dfOpen

df["Date"] = pd.to_datetime(df["Date"])
df["Tanggal"] = df["Date"].dt.date
df["Hari"] = df["Date"].dt.day_name()

df = df[["Date","Hari","Tanggal"] + kolomBaruSum]

df["Sum"] = df[kolomBaruSum].sum(axis=1)

df = df[["Date", "Hari", "Tanggal", "Sum"]]
df = df.groupby(["Tanggal", "Hari"])["Sum"].sum().reset_index()
```

```python
df
```

```python
plt.figure(figsize=(16,12))

sns.boxplot(df, x = "Hari", y = "Sum")
plt.show()
```
